```
def busqueda (Producto):
    #lista de sucursales excluidas

    dfBuscado = dfStock.filter((dfStock.Producto == Producto) & (dfStock.Stock > 5)).select('Sucursal')
    
    if (dfBuscado.rdd.isEmpty()):
        dfStockPanda = dfBuscado.toPandas()
        listasSucursalesExcluidas = dfStockPanda.to_numpy().transpose().tolist()
        listaSucursalesExcluidas = [i for i in listasSucursalesExcluidas[0]]
        display(dfBuscado)
        return False
    else:
        return True
```

```
def ordenarProductoPorVentas(Producto, cantidadVentas):
    for i in range (1,10):
        if i in listaSucursalesExcluidas:
            continue
        else:
            dfVentasInternetOrdenado = dfVentasInternet.filter((dfVentasInternet.Producto == Producto) & (dfVentasInternet.Cod_Sucursal == i))
            
            if (dfVentasInternetOrdenado.rdd.isEmpty()):
                continue
            else:
                cantidadVendida = dfVentasInternetOrdenado.select(sum('Cantidad'))
                cantidad = cantidadVendida.first()['sum(Cantidad)']

                # Realizo la union con los datos ingresados
                newRow = spark.createDataFrame([(Producto, cantidad, i)])
                cantidadVentas = cantidadVentas.union(newRow)
                
    cantidadVentas = cantidadVentas.sort(col("Cantidad").desc())
    
    display(cantidadVentas)
```

```
def crearDataFrame():
    # Creo un nuevo data frame
    # Import necessary libraries

    # Create a spark session
    spark = SparkSession.builder.appName('Empty_Dataframe').getOrCreate()

    # Create an empty RDD
    emp_RDD = spark.sparkContext.emptyRDD()

    # Create an expected schema
    columns = StructType([StructField('Producto',
                                      StringType(), True),
                        StructField('Cantidad',
                                    IntegerType(), True),
                        StructField('Sucursal',
                                    IntegerType(), True)])

    # Create an empty RDD with expected schema
    dfCantidadVentas = spark.createDataFrame(data = emp_RDD,
                               schema = columns)
    return dfCantidadVentas
#     # Print the dataframe
#     print('Dataframe :')
#     dfCantidadVentas.show()

#     # Print the schema
#     print('Schema :')
#     dfCantidadVentas.printSchema()
```

```
def buscarCategoria(Categoria):
    dfCategorias = dfStock.filter((dfStock.Categoria == Categoria))
    if(dfCategorias.rdd.isEmpty()):
        codCategoria = dfStock.select([max("Cod_Categoria")])
        dfArrayCategorias = codCategoria.toPandas()
        categoria_array = dfArrayCategorias.to_numpy().transpose().tolist()
        categoria_arrays = [i for i in categoria_array[0]]
        categoria_arrays[0] = categoria_arrays[0] + 1
        return categoria_arrays[0]
    else:
        codCategoria = dfCategorias.select('Cod_Categoria')
        Cod_Categoria = codCategoria.first()['Cod_Categoria']
        return Cod_Categoria
```

```
def buscarSubCategoria(SubCategoria):
    dfSubCategorias = dfStock.filter((dfStock.SubCategoria == SubCategoria))
    if(dfSubCategorias.rdd.isEmpty()):
        codSubCategoria = dfStock.select([max("Cod_SubCategoria")] )
        dfArraysubCategorias = codSubCategoria.toPandas()
        subcategoria_array = dfArraysubCategorias.to_numpy().transpose().tolist()
        subcategoria_arrays = [i for i in subcategoria_array[0]]
        subcategoria_arrays[0] = subcategoria_arrays[0] + 1 
        return subcategoria_arrays[0]
    else:
        codSubCategoria = dfSubCategorias.select('Cod_SubCategoria')
        cod_SubCategoria = codSubCategoria.first()['Cod_SubCategoria']
        return cod_SubCategoria
    
```

```
def buscarProducto ():
        codProducto = dfStock.select([max("Cod_Producto")])
        dfArraycodProducto = codProducto.toPandas()
        codProducto_array = dfArraycodProducto.to_numpy().transpose().tolist()
        codProducto_arrays = [i for i in codProducto_array[0]]
        codProducto_arrays[0] = codProducto_arrays[0] + 1 
        return codProducto_arrays[0]
```

```
def export_csv(df, filePath, fileName):
    temp_path = f"{filePath}/__temp"
    target_path = f"{filePath}/{fileName}.csv"

    df.coalesce(1).write.mode("overwrite").option("header",True).csv(temp_path)

    Path = sc._gateway.jvm.org.apache.hadoop.fs.Path

    # get the part file generated by spark write
    fs = Path(temp_path).getFileSystem(sc._jsc.hadoopConfiguration())
    csv_part_file = fs.globStatus(Path(temp_path + "/part*"))[0].getPath()

    # move and rename the file
    fs.rename(csv_part_file, Path(target_path))
    fs.delete(Path(temp_path), True)
```
